````python
size = int(input("Please enter the table size: "))
# Print a size x size multiplication table
# First, print heading: 1 2 3 4 5 etc.
print(" ", end='')
# Print column heading
for column in range(1, size + 1):
  print('{0:4}'.format(column), end='') # Display column number
print() # Go down to the next line
# Print line separator: +------------------
print(" +", end='')
for column in range(1, size + 1):
  print('----', end='') # Display line
print() # Drop down to next line

# Print table contents
for row in range(1, size + 1):
  print('{0:3} |'.format(row), end='') # Print heading for this row
  for column in range(1, size + 1):
    product = row*column # Compute product
    print('{0:4}'.format(product), end='') # Display product
  print()
````

````python
#Probar debugger en VSCode
max_value = int(input('Display primes up to what value? '))
value = 2 # Smallest prime number
while value <= max_value:
  # See if value is prime
  is_prime = True # Provisionally, value is prime
  # Try all possible factors from 2 to value - 1
  trial_factor = 2
  while trial_factor < value:
    if value % trial_factor == 0:
      is_prime = False # Found a factor
      break # No need to continue; it is NOT prime
    trial_factor += 1 # Try the next potential factor
  if is_prime:
    print(value, end= ' ') # Display the prime number
  value += 1 # Try the next potential prime number
print() # Move cursor down to next line
````

````python
class Animal:
    def __init__(self, nombre, especie):
        self.nombre = nombre
        self.especie = especie

    def hacer_sonido(self):
        return "El animal hace un sonido."

    def describir(self):
        return f"{self.nombre} es un {self.especie}."

class Perro(Animal): # Clase Perro hereda de la clase Animal
    def __init__(self, nombre, raza):
        super().__init__(nombre, "Perro")  # Llama al constructor de la clase base
        self.raza = raza

    def hacer_sonido(self):
        return "¡Guau!"

    def describir(self):
        return f"{self.nombre} es un {self.raza} y es un {self.especie}."

# Ejemplo de uso
if __name__ == "__main__":
    mi_perro = Perro("Fido", "Labrador")

    print(mi_perro.describir())  # Uso del método sobreescrito
    print(mi_perro.hacer_sonido())  # Uso del método sobreescrito
````

# Herencia múltiple

````python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def reproducirse(self):
        return "Me reproduzco, es una de mis funciones vitales "


class Humano:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def hablar(self):
        return "Hola, soy un hablador."


class Estudiante(Animal, Humano):
    def __init__(self, nombre, edad, carrera):
        Animal.__init__(self, nombre)  # Inicializa la clase Animal
        Humano.__init__(self, nombre, edad)  # Inicializa la clase Humano
        self.carrera = carrera

    def estudiar(self):
        return f"{self.nombre} está estudiando {self.carrera}."


# Ejemplo de uso
if __name__ == "__main__":
    estudiante = Estudiante("Juan", 20, "Ingeniería")

    print(estudiante.reproducirse())  # Llama al método de Animal
    print(estudiante.hablar())          # Llama al método de Humano
    print(estudiante.estudiar())        # Llama al método específico de Estudiante
````

````python

class Humano:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    def hablar(self):
        return "I speak English and Spanish."

class Parrandero:
    def __init__(self, nombre):
        self.nombre = nombre

    def Celebrar(self):
        return "Bebiendo todo el día, de fiesta vivo YO "

class Estudiante(Parrandero, Humano):
    def __init__(self, nombre, edad, carrera):
        Parrandero.__init__(self, nombre)  # Inicializa la clase Animal
        Humano.__init__(self, nombre, edad)  # Inicializa la clase Humano
        self.carrera = carrera

    def estudiar(self):
        return f"{self.nombre} está estudiando {self.carrera}."


# Ejemplo de uso
if __name__ == "__main__":
    estudiante = Estudiante("Juan", 20, "Ingeniería")

    print(estudiante.Celebrar())  # Llama al método de Animal
    print(estudiante.hablar())          # Llama al método de Humano
    print(estudiante.estudiar())        # Llama al método específico de Estudiante
````

# Polimorfimso

 Ocurre cuando una clase que hereda de otra sobrescribe métodos de la clase padre y los personaliza a su necesidad
  En el ejemplo siguente aunque estamos llamando al método hacer_sonido en objetos de tipo Animal, el polimorfismo permite que cada clase hija (Perro y Gato) tenga su propia implementación del método. Eso es lo que hace que el perro diga "Guau" y el gato diga "Miau".

````python
class Animal:
    def hacer_sonido(self):
        pass  # Método vacío para ser sobrescrito en las clases hijas
````

````python
class Perro(Animal):
    def hacer_sonido(self):
        return "Guau"

class Gato(Animal):
    def hacer_sonido(self):
        return "Miau"
````

````python
# Lista de animales
animales = [Perro(), Gato()]

# Polimorfismo: aunque todos son 'Animal', el método 'hacer_sonido' se comporta de manera diferente dependiendo del tipo de objeto o clase
for animal in animales:
    print(animal.hacer_sonido())
````

# **Relaciones entre clases**
##  - **1. Asociación**

### Es la relación más general: un objeto usa a otro, sin que haya propiedad ni dependencia fuerte.
**"Un objeto conoce a otro"**

````python
class Estudiante:
    def __init__(self, nombre):
        self.nombre = nombre

class Curso:
    def __init__(self, nombre):
        self.nombre = nombre
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)

# Ejemplo de uso
curso = Curso("Matemáticas")
estudiante = Estudiante("Juan")
curso.agregar_estudiante(estudiante)
````

## **2. Agregación**

La agregación es similar a la composición, pero más **flexible**: el objeto agregado no depende completamente de la clase que lo contiene **(relación más "débil")**.

````python
class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre

class Equipo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.jugadores = []

    def agregar_jugador(self, jugador):
        self.jugadores.append(jugador)

# Ejemplo de uso
equipo = Equipo("Fútbol")
jugador1 = Jugador("Pedro")
equipo.agregar_jugador(jugador1)

print(equipo.jugadores[0].nombre)  # Salida: Pedro
````

````python
class Profesor:
    def __init__(self, nombre):
        self.nombre = nombre

class Curso:
    def __init__(self, nombre, profesor):
        self.nombre = nombre
        self.profesor = profesor  # Agregación: se pasa desde fuera

prof = Profesor("Dr. Pérez")
curso = Curso("Matemáticas", prof)

print(curso.profesor.nombre)  # ✔️ Comunicación entre objetos
````

````python
class Monitor:
    def __init__(self, modelo):
        self.modelo = modelo

class Computadora:
    def __init__(self, monitor):
        self.monitor = monitor  # Se pasa desde afuera

monitor1 = Monitor("LG 24'")
pc = Computadora(monitor1)
````

## **3. Composición**

La composición se presenta cuando una clase contiene a otra clase como parte esencial de su estructura. Es una relación **"tiene un"**

````python
class Habitacion:
    def __init__(self, nombre):
        self.nombre = nombre

class Casa:
    def __init__(self):
        self.habitaciones = []
    def __str__(self):
        return f"Casa con {len(self.habitaciones)} habitaciones"

    def agregar_habitacion(self, nombre):
        habitacion = Habitacion(nombre)
        self.habitaciones.append(habitacion)

# Ejemplo de uso
casa = Casa()
casa.agregar_habitacion("Sala")
casa.agregar_habitacion('Comedor')
print(casa)  # Salida: Casa con 2 habitaciones
````

````python
class Motor:
    def arrancar(self):
        print("Motor arrancado")

class Auto:
    def __init__(self):
        self.motor = Motor()  # Composición: el Auto tiene un Motor

    def encender(self):
        self.motor.arrancar()

mi_auto = Auto()
mi_auto.encender()
````

````python
class Computadora:
    def __init__(self):
        self.monitor = Monitor("LG 24'")  # Se crea internamente

class Monitor:
    def __init__(self, modelo):
        self.modelo = modelo

pc = Computadora()
````

## 4. Herencia

````python
class Animal:
    def hacer_sonido(self):
        pass

class Perro(Animal):
    def hacer_sonido(self):
        return "Guau"

class Gato(Animal):
    def hacer_sonido(self):
        return "Miau"

# Ejemplo de uso
mi_perro = Perro()
print(mi_perro.hacer_sonido())  # Salida: Guau
````

## 5. Dependencia

 Un objeto usa temporalmente a otro, como parámetro en un método, pero no lo guarda.

**"Depende de" pero solo por un momento**

````python
class Impresora:
    def imprimir(self, documento):
        print(f"Imprimiendo {documento.contenido}")

class Documento:
    def __init__(self, contenido):
        self.contenido = contenido

I1 = Impresora()
D1 = Documento("Bienvenidos a las relaciones entre clases")
I1.imprimir(D1)
````

## Relaciones de Muchos a Muchos entre clases

````python
class Estudiante:
    def __init__(self, nombre):
        self.nombre = nombre
        self.cursos = []

    def agregar_curso(self, curso):
        if curso not in self.cursos:
            self.cursos.append(curso)
            curso.agregar_estudiante(self)

class Curso:
    def __init__(self, nombre):
        self.nombre = nombre
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        if estudiante not in self.estudiantes:
            self.estudiantes.append(estudiante)
            estudiante.agregar_curso(self)

# Crear instancias de estudiantes y cursos
estudiante1 = Estudiante("Juan")
estudiante2 = Estudiante("Ana")
curso1 = Curso("Matemáticas")
curso2 = Curso("Historia")

# Establecer relaciones de muchos a muchos
estudiante1.agregar_curso(curso1)
estudiante1.agregar_curso(curso2)
estudiante2.agregar_curso(curso1)

# Verificar las relaciones
print(f"{estudiante1.nombre} está inscrito en: {[curso.nombre for curso in estudiante1.cursos]}")
print(f"{estudiante2.nombre} está inscrito en: {[curso.nombre for curso in estudiante2.cursos]}")
print(f"{curso1.nombre} tiene inscritos a: {[estudiante.nombre for estudiante in curso1.estudiantes]}")
print(f"{curso2.nombre} tiene inscritos a: {[estudiante.nombre for estudiante in curso2.estudiantes]}")
````

## Rompiendo la relacion de muchos a muchos con una clase intermedia

````python
class Inscripcion:
    def __init__(self, estudiante, curso):
        self.estudiante = estudiante
        self.curso = curso

class Estudiante:
    def __init__(self, nombre):
        self.nombre = nombre
        self.inscripciones = []

    def agregar_curso(self, curso):
        inscripcion = Inscripcion(self, curso)
        self.inscripciones.append(inscripcion)
        curso.inscripciones.append(inscripcion)

class Curso:
    def __init__(self, nombre):
        self.nombre = nombre
        self.inscripciones = []

# Crear instancias de estudiantes y cursos
estudiante1 = Estudiante("Juan")
estudiante2 = Estudiante("Ana")
curso1 = Curso("Matemáticas")
curso2 = Curso("Historia")

# Establecer relaciones de muchos a muchos
estudiante1.agregar_curso(curso1)
estudiante1.agregar_curso(curso2)
estudiante2.agregar_curso(curso1)

# Verificar las relaciones
for inscripcion in estudiante1.inscripciones:
    print(f"{estudiante1.nombre} está inscrito en {inscripcion.curso.nombre}")

for inscripcion in curso1.inscripciones:
    print(f"{curso1.nombre} tiene inscrito a {inscripcion.estudiante.nombre}")
````
